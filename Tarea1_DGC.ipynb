## Primer punto Distribuciones discretas.

```python
#Para todas las simulaciones:
N=100000 # Tamaño de muestras
m=3000  # Anchos de secuencias
numpy.random.seed(seed=2**32 - 1)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# bernoulli con cierto p entre 0 y 1
p = 0.3
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
print ("Media: ",mean, "Varianza: ",var, "F1: ",skew, "F2: ",kurt)
r = bernoulli.rvs(p, size=N) #Generate random numbers
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
```

```python
r.sort() #Ordenar
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli ordenados con p="+str(p))
plt.show()
```

```python
plt.hist(r, bins = m) #Histograma, "r" son los números aleatorios de bernoulli y "m" son los anchos iguales de bins 
plt.title("Histograma de los "+str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean() #Media de los números aleatorios
varest=r.var() #La varianza de los números aleatorios
print("Los valores teoricos de la media y varianza para la bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

## Segundo punto Distribuciones contínuas.


```python
#Para todas las simulaciones:
N=1000
m=300
numpy.random.seed(seed=1314926535)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# Normal de media 5 y desviación estandar 0.5
u=5
std=0.5
mean, var, skew, kurt = scipy.stats.norm.stats(moments='mvsk',loc=u,scale=std)
r = scipy.stats.norm.rvs (loc=u,scale=std, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales con media="+str(u)+" y std="+str(std))
plt.show()
r.sort()
plt.plot(r)
#plt.title(str(N)+" puntos distribuidos normales ordenados con p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos normales con u="+str(u)+ " y std= "+str(std))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```python
# Uniforme entre 0 y 255
inicio = 0
fin = 255
from scipy.stats import uniform
mean, var, skew, kurt = uniform.stats(moments='mvsk',loc=inicio,scale=fin)
r = scipy.stats.uniform.rvs (loc=inicio, scale=fin, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos de manera uniforme entre "+str(inicio)+" y "+str(fin))
plt.show()
r.sort()
plt.plot(r)
#plt.title(str(N)+" puntos distribuidos de manera uniforme ordenados con p: "+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos uniformes entre "+str(inicio)+ " y "+str(fin))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avgu=r.mean()
varestu=r.var()
print("Los valores teoricos de la media y varianza para la uniforme son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avgu,"var_est=",varestu)
```

```python
# Exponencial con Lamnda L
L=0.3
inicio = 0
fin = 1/L

from scipy.stats import expon
mean, var, skew, kurt = expon.stats(moments='mvsk', loc=inicio, scale=fin)
r = scipy.stats.expon.rvs (loc=inicio, scale=fin, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html

plt.plot(r)
plt.title(str(N)+" puntos distribuidos de manera exponencial con Lamnda igual a"+str(L))
plt.show()
r.sort()
plt.plot(r)
#plt.title(str(N)+" puntos distribuidos de manera exponencial ordenados con p: "+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos de manera exponencial con Lamnda "+str(L))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avgex=r.mean()
varestex=r.var()
print("Los valores teoricos de la media y varianza para la exponencial son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avgex,"var_est=",varestex)
```

```python
#Chi cuadrada con K grados de libertad
K=3.0

from scipy.stats import chi2
mean, var, skew, kurt = chi2.stats(K, moments='mvsk')
rc = scipy.stats.chi2.rvs (K, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html

plt.plot(rc)
plt.title(str(N)+" puntos distribuidos de manera Chi cuadrado con K: "+str(K)+ " grados de libertad")
plt.show()
rc.sort()
plt.plot(rc)

#plt.title(str(N)+" puntos distribuidos de manera chi cuadrado)
plt.show()
plt.hist(rc, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos de manera Chi cuadrado con K: "+str(K)+ " grados de libertad")
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo rc:")
print(type(rc))
print("Vamos a hallar algunas estimaciones de estadisticas de rc")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avgc=rc.mean()
varestc=rc.var()
print("Los valores teoricos de la media y varianza para la forma chi cuadrado son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avgc,"var_est=",varestc)
```
